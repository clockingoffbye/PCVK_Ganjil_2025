Tugas 2 Tampilkan image dalam channel Red-Blue dan Green-Blue saja!

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

img_path = '/content/drive/MyDrive/PCV_Wk1_Images/kitten01.jpg'
img = cv2.imread(img_path)  # BGR
b, g, r = cv2.split(img)  # Split channel: B, G, R

# Red-Blue: Gabung R dan B (abaikan G)
rb = cv2.merge([b, np.zeros_like(g), r])
cv2_imshow(rb)

# Green-Blue: Gabung G dan B (abaikan R)
gb = cv2.merge([b, g, np.zeros_like(r)])
cv2_imshow(gb)
```

Tugas 4 Tampilkan image baris ke 5-30, semua kolom, channel Red saja!

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

img_path = '/content/drive/MyDrive/PCV_Wk1_Images/kitten01.jpg'
img = cv2.imread(img_path)  # BGR

# Crop baris 5:30, semua kolom
cropped = img[5:30, :, :]
b, g, r = cv2.split(cropped)  # Split

# Channel Red saja (R di BGR adalah index 2)
red_channel = cv2.merge([np.zeros_like(b), np.zeros_like(g), r])
cv2_imshow(red_channel)
```

Tugas 5 Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. Disarankan menggunakan bilangan acak/random!

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Buat black image 512x512, 3 channel
img = np.zeros((512, 512, 3), dtype=np.uint8)

# Random ukuran dan warna (seed untuk reproducible)
np.random.seed(42)
colors = [(255, 0, 0), (0, 255, 0), (0, 0, 255), (255, 255, 0), (255, 0, 255)]  # BGR colors
for i in range(5):
    x, y = np.random.randint(50, 400, 2)  # Posisi random
    w, h = np.random.randint(50, 150, 2)  # Ukuran random
    cv2.rectangle(img, (x, y), (x+w, y+h), colors[i], -1)  # Filled rectangle

cv2_imshow(img)
```

Tugas 7 Buat rectangle dan circle pada bagian wajah dari image foto anda saat beraktifitas (bukan pasfoto).

```python
import cv2
from google.colab.patches import cv2_imshow

img_path = '/content/drive/MyDrive/PCV_Wk1_Images/my_photo.jpg'  # Ganti dengan foto Anda
img = cv2.imread(img_path)

# Rectangle di wajah (ganti koordinat: (x1,y1), (x2,y2))
cv2.rectangle(img, (100, 100), (200, 200), (0, 255, 0), 2)  # Hijau, tebal 2

# Circle di wajah (center (x,y), radius)
cv2.circle(img, (150, 150), 50, (255, 0, 0), 2)  # Biru, tebal 2

cv2_imshow(img)
```

Tugas 8 Buat rectangle pada bagian sudut bawah kiri channel B pada color space RGB dari citra kitten/ lena/ mandrill/ male/ female/ couple/ sailboat/ peppers!

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

img_path = '/content/drive/MyDrive/PCV_Wk1_Images/lena.png'
img_bgr = cv2.imread(img_path)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Ambil channel B
b_channel = img_rgb[:, :, 0].copy()   # <-- fix: pakai copy()

# Ukuran gambar
h, w = b_channel.shape

# Rectangle putih di bawah kiri
cv2.rectangle(b_channel, (0, h-100), (100, h), 255, -1)

# Gabungkan kembali
img_rgb[:, :, 0] = b_channel

# Tampilkan
cv2_imshow(cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR))
```

Tugas 9 Lengkapi tulisan nama file pada file citra dari soal no.8. Gunakan font, ukuran font, dan warna font yang sesuai keinginan anda.

```python
import cv2
from google.colab.patches import cv2_imshow

# Lanjut dari img_rgb di tugas 8, convert ke BGR untuk text
img_bgr = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR)

# Text: nama file, posisi sudut kiri bawah, font HERSHEY_SIMPLEX (ukuran 1), tebal 2, warna merah (0,0,255 BGR)
cv2.putText(img_bgr, 'lena.jpg', (10, img_bgr.shape[0]-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

cv2_imshow(img_bgr)
```
